```python
### PLOT LEARNING CURVES ###
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pickle

# Fetch saved loss dictionaries
f = open('lr3e-4_loss.pkl','rb')
bs1 = pickle.load(f)
f.close()
f = open('bs4_loss.pkl','rb')
bs4 = pickle.load(f)
f.close()
f = open('bs8_loss.pkl','rb')
bs8 = pickle.load(f)
f.close()
mpl.rcParams['figure.dpi']= 500
x = np.arange(50)

### BATCH SIZE TRAINING ###
plt.plot(x,bs1['train'],label='Batch Size = 1')
plt.plot(x,bs4['train'],label='Batch Size = 4')
plt.plot(x,bs8['train'],label='Batch Size = 8')
plt.legend()
plt.title('Train Dose Score with Learning Rate = 0.0003')
plt.xlabel('Epoch')
plt.ylabel('Dose Score (MAE)')
plt.show()
```

```python
### BATCH SIZE VAL ###
plt.plot(x,bs1['val'],label='Batch Size = 1')
plt.plot(x,bs4['val'],label='Batch Size = 4')
plt.plot(x,bs8['val'],label='Batch Size = 8')
plt.legend()
plt.title('Val Dose Score with Learning Rate = 0.0003')
plt.xlabel('Epoch')
plt.ylabel('Dose Score (MAE)')
plt.show()
```

```python
### LR TRAINING ###
f = open('lr3e-4_loss.pkl','rb')
lr3e4 = pickle.load(f)
f.close()
f = open('lr1e-2_loss.pkl','rb')
lr1e2 = pickle.load(f)
f.close()
f = open('lr1e-4_loss.pkl','rb')
lr1e4 = pickle.load(f)
f.close()

plt.plot(x,lr1e4['train'],label='Learning Rate = 0.0001')
plt.plot(x,lr3e4['train'],label='Learning Rate = 0.0003')
plt.plot(x,lr1e2['train'],label='Learning Rate = 0.01')
plt.legend()
plt.title('Train Dose Score with Batch Size = 1')
plt.xlabel('Epoch')
plt.ylabel('Dose Score (MAE)')
plt.show()
```

```python
### LR VAL ###
plt.plot(x,lr1e4['val'],label='Learning Rate = 0.0001')
plt.plot(x,lr3e4['val'],label='Learning Rate = 0.0003')
plt.plot(x,lr1e2['val'],label='Learning Rate = 0.01')
plt.legend()
plt.title('Val Dose Score with Batch Size = 1')
plt.xlabel('Epoch')
plt.ylabel('Dose Score (MAE)')
plt.show()
```
